```python
import matplotlib.pyplot as plt
import numpy as np
```

## 基础应用

```python
x = np.linspace(-1,1,50)
y1 = 2*x + 1
y2 = x**2
```

```python
#Figure size & number
plt.figure(num=1,figsize=(10,5))
# Plot parameter
plt.plot(x,y1,linewidth=1.0)
plt.plot(x,y2,color='red',linewidth=1.0,linestyle='--')
#X-limit and Y-limit
plt.xlim((-1,1))
plt.ylim((-2,4))
#X-lable and Y-label
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
